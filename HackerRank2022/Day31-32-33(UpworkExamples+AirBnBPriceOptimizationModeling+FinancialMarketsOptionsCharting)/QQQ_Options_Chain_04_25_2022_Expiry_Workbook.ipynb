```python
import pandas as pd
import numpy as np
```

```python
# 04/18/2022 qqq puts/calls open interest
QQQ_Options_Chain_04_25_2022_Expiry_As_Of_04_22_2022 = pd.read_csv(r'QQQ_Options_Chain_04_25_2022_Expiry(As_Of_04_22_2022).csv')
```

```python
QQQ_Options_Chain_04_25_2022_Expiry_As_Of_04_22_2022.head()
```

```python
# 04/18/2022 qqq puts/calls open interest
QQQ_Options_Chain_04_25_2022_Expiry_As_Of_04_22_2022 = QQQ_Options_Chain_04_25_2022_Expiry_As_Of_04_22_2022[['CallVolume','CallOpenInt','Strike','PutVolume','PutOpenInt']]
```

```python
# 04/18/2022 qqq puts/calls open interest
QQQ_Options_Chain_04_25_2022_Expiry_As_Of_04_22_2022
```

```python
QQQ_Options_Chain_04_25_2022_Expiry_As_Of_04_22_2022_copy = QQQ_Options_Chain_04_25_2022_Expiry_As_Of_04_22_2022.copy()
```

```python
QQQ_Options_Chain_04_25_2022_Expiry_As_Of_04_22_2022_copy
```

```python
# 04/18/2022 qqq puts/calls open interest
calls = QQQ_Options_Chain_04_25_2022_Expiry_As_Of_04_22_2022['CallOpenInt']
puts = QQQ_Options_Chain_04_25_2022_Expiry_As_Of_04_22_2022['PutOpenInt']
strikes = QQQ_Options_Chain_04_25_2022_Expiry_As_Of_04_22_2022['Strike']

strikes = [x for x in QQQ_Options_Chain_04_25_2022_Expiry_As_Of_04_22_2022['Strike']]
len(strikes)
```

```python
# 04/18/2022 qqq puts/calls open interest
callsX = []
for x in QQQ_Options_Chain_04_25_2022_Expiry_As_Of_04_22_2022['Strike']:
    x = x +.4
    callsX.append(x)
print(callsX)
```

```python
# 04/18/2022 qqq puts/calls open interest
import matplotlib.pyplot as plt
# Data to be used for bars
data = [puts,
calls]
# Strikes for X axis
X = strikes
# Create a new figure
fig = plt.figure()

ax = fig.add_axes([0,0,20,10])
plt.xticks(fontsize=60)
plt.yticks(fontsize=100)
plt.xticks(strikes)
parameters = {'axes.titlesize': 100}
plt.rcParams.update(parameters)
plt.title('04/25/2022 QQQ Puts/Calls Open Interest')
line1 = ax.bar(X, data[0], color = 'r', width = 0.25, label='Puts')
line2 = ax.bar(callsX, data[1], color = 'g', width = 0.25, label='Calls')
# ax.bar(X, data[0], color = 'b', width = 0.25)
# ax.bar(callsX, data[1], color = 'g', width = 0.25)
# ax.legend(handles = [line1, line2], bbox_to_anchor=(0.35, 1.0), borderpad=50, prop={"size":50})
ax.legend(handles = [line1, line2], borderpad=2, prop={"size":100})
```
