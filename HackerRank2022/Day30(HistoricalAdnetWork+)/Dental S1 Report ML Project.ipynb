```python
# Each column per date range = own list (high rpc words, average rpc breakpoint, low rpc words)
# For each word +1/-1 per space higher/lower than the average RPC
# For each word in RPC column, find match in Volume if exists
# If no match in RPC/Volume per word in Volume/RPC, trash word from running
# For each word in Volume column, highest word = 1.00, each slot beneath highest word incrementally decreases by equal amount
#  until .1
# Therefore, words at the bottom of the RPC will have large negative modifiers and words at the bottom of volume will have 
#  leading to: large negative value for low RPC, high volume matches; large positive values for high RPC, high volume matches;
#  small positive values for high RPC, low volume; small negative value for low RPC low volume
```
